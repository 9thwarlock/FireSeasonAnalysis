## <ins>2018 in Review</ins>

The wildfires in 2018 were record breaking in regards to the amount of damage to property. In November, Diablo winds in Northern California formed a wave of wildfires as well as spread already existing fires, which included Woolsey fire and the Camp Fire. Camp Fire destroyed more than 18,000 structures, therefore becoming one of the deadliest wildfires in the state's history. The Mendocino Complex fire in Southern California burned half a million acres, also settings a record of becoming the largest complex fire in California history.

```python
import requests
import pandas
import folium

df = pandas.read_csv("mapdataall.csv")
# Filter rows based on condition
df = df[df['incident_date_created'].str.startswith('2018')]

# Create a map centered on the first matching incident's location
#map_center = [df['incident_latitude'].iloc[0], df['incident_longitude'].iloc[0]]
#m = folium.Map(location=map_center, zoom_start=12)
m = folium.Map(location=[37.3, -119.4], zoom_start=6)

# Iterate over matching rows and add markers to the map
for i, row in df.iterrows():
    location = [row['incident_latitude'], row['incident_longitude']]
    tooltip = f"Incident {i}: {row['incident_name']}"
    url = row['incident_url']
    folium.Marker(location=location,popup=url, tooltip=tooltip).add_to(m)

# Display the map
m
```
