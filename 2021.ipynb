## <ins>2021 in Review</ins>

With 2021 the most notable aspect of the wildfire season was its incredibly early start. In January there were already 297 fires burning, in part due to the drought and low rainfall levels. 1,171 acres of land burned, which set a new record for destruction in January as the damage was 20 times the 5 year average for January. Just as the previous year, August proved to be especially difficult as the Caldor & Dixie fires wreaked havoc and destroyed thousands of buildings. Fortunately rainfall in October helped assist with remaining fires and reduced wildfire risk statewide.

```python
import requests
import pandas
import folium

df = pandas.read_csv("mapdataall.csv")
# Filter rows based on condition
df = df[df['incident_date_created'].str.startswith('2021')]

# Create a map centered on the first matching incident's location
#map_center = [df['incident_latitude'].iloc[0], df['incident_longitude'].iloc[0]]
#m = folium.Map(location=map_center, zoom_start=12)
m = folium.Map(location=[37.3, -119.4], zoom_start=6)

# Iterate over matching rows and add markers to the map
for i, row in df.iterrows():
    location = [row['incident_latitude'], row['incident_longitude']]
    tooltip = f"Incident {i}: {row['incident_name']}"
    url = row['incident_url']
    folium.Marker(location=location,popup=url, tooltip=tooltip).add_to(m)

# Display the map
m
```
