## <ins>2019 in Review</ins>

When compared to 2017 and 2018, 2019 was a calmer year for wildfires, yet still destructive. According to fire.ca.gov, The Kincade Fire was the largest fire of the year as 77,758 acres in Sonoma were burned. A notable event during the year's fires season was the shutting off of power to 800,000 households to reduce the risk of possible wildfires due to electrical arcing in high winds from power lines.

```python
import requests
import pandas
import folium

df = pandas.read_csv("mapdataall.csv")
# Filter rows based on condition
df = df[df['incident_date_created'].str.startswith('2019')]

# Create a map centered on the first matching incident's location
#map_center = [df['incident_latitude'].iloc[0], df['incident_longitude'].iloc[0]]
#m = folium.Map(location=map_center, zoom_start=12)
m = folium.Map(location=[37.3, -119.4], zoom_start=6)

# Iterate over matching rows and add markers to the map
for i, row in df.iterrows():
    location = [row['incident_latitude'], row['incident_longitude']]
    tooltip = f"Incident {i}: {row['incident_name']}"
    url = row['incident_url']
    folium.Marker(location=location,popup=url, tooltip=tooltip).add_to(m)

# Display the map
m
```
