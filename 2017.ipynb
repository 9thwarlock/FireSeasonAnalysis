## <ins>2017 in Review</ins>

2017 was an especially destructive year as 230,000 people were forced to evacuate. This year at the time marked the second hottest year in California history. The high temperatures meant that the landscape was incredibly dry, contributing to high amount of wildfires and acres burned. While the California Fire Season typically ends around September, Santa Ana winds in December sparked a new wave of fires, include the Thomas Fire, which at the time was the largest wildfire in Southern California.

```python
import requests
import pandas
import folium

df = pandas.read_csv("mapdataall.csv")
# Filter rows based on condition
df = df[df['incident_date_created'].str.startswith('2017')]

# Create a map centered on the first matching incident's location
#map_center = [df['incident_latitude'].iloc[0], df['incident_longitude'].iloc[0]]
#m = folium.Map(location=map_center, zoom_start=12)
m = folium.Map(location=[37.3, -119.4], zoom_start=6)

# Iterate over matching rows and add markers to the map
for i, row in df.iterrows():
    location = [row['incident_latitude'], row['incident_longitude']]
    tooltip = f"Incident {i}: {row['incident_name']}"
    url = row['incident_url']
    folium.Marker(location=location,popup=url, tooltip=tooltip).add_to(m)

# Display the map
m
```
