## <ins>2020 in Review</ins>

With the burning of 4.3 million acres, 2020 was the largest wildfire season recorded in California's history. As detailed in the fire.ca.gov coverage of the year's fires, the August Complex fire was deemed as the first "gigafire" as the wildfire covered seven different counties and "became larger that the state of Rhode Island". By August of 2020 there were 367 concurrent fires throughout the state, and a combination of Diablo & Santa Ana winds in September further spread the active fires.

```python
import requests
import pandas
import folium

df = pandas.read_csv("mapdataall.csv")
# Filter rows based on condition
df = df[df['incident_date_created'].str.startswith('2020')]

# Create a map centered on the first matching incident's location
#map_center = [df['incident_latitude'].iloc[0], df['incident_longitude'].iloc[0]]
#m = folium.Map(location=map_center, zoom_start=12)
m = folium.Map(location=[37.3, -119.4], zoom_start=6)

# Iterate over matching rows and add markers to the map
for i, row in df.iterrows():
    location = [row['incident_latitude'], row['incident_longitude']]
    tooltip = f"Incident {i}: {row['incident_name']}"
    url = row['incident_url']
    folium.Marker(location=location,popup=url, tooltip=tooltip).add_to(m)

# Display the map
m
```
